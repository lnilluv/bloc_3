```python
import datetime
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import average_precision_score
from catboost import CatBoostClassifier, Pool

print(" CATBOOST ".center(80, "="))

print("Loading dataset...")

df = pd.read_csv("C:/Users/laure/PycharmProjects/pythonProject/conversion_data_train.csv")

X = df.drop(["converted"], axis=1)
y = df["converted"]

# X = X.select_dtypes(include=["int64", "float64"])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)

categoricals = ["country", "new_user", "source"]
# numericals = ["age", "total_pages_visited"]

# model = CatBoostClassifier(iterations=2, depth=2, learning_rate=1, loss_function='Logloss', cat_features=categoricals, verbose=True)

model = CatBoostClassifier(iterations=10000, learning_rate=0.1, depth=10, rsm=1, loss_function="Logloss", eval_metric="F1", cat_features=categoricals, verbose=500,  task_type="GPU", random_seed = 42, devices='0:1')

# print("Training partial model")

# model.fit(X_train, y_train)

# y_pred = model.predict_proba(X_test)
# y_pred_class1 = np.array(y_pred)[:, -1]

# print(average_precision_score(y_test, y_pred_class1))

print("Training full model")

model.fit(X_train, y_train)

model.save_model("miaoutwo")

# model.load_model("miaou")

df_test = pd.read_csv("C:/Users/laure/PycharmProjects/pythonProject/conversion_data_test.csv")

predictions = model.predict(df_test)
print(predictions)

# predictions_class1 = np.array(predictions)[:, -1]
# print(predictions_class1)

pred_df = pd.DataFrame(columns=['converted'],data=predictions)
# pred_df = pd.DataFrame({ "converted": np.array(predictions)[:, -1] })
pred_df.to_csv(f"submission-catboost-{datetime.datetime.now().timestamp()}.csv", index=True, header=True)
```
